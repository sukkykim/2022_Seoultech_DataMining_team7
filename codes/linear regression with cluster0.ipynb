```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

cluster0 인 data를 test 와 val와  train으로 나누기
train : 17년도 1분기 ~ 21년도 2분기
val : 21년도 3분기
test : 21년도 4분기 

```python
dataset1=pd.read_csv('cluster_Label0.csv',index_col=0)
dataset1
```

```python
X_dataset=dataset1.drop(columns=['year','session','total_sale','s_zone','s_zone_code','s_zone_name'])
y_dataset=dataset1['total_sale']
X_test=X_dataset[:400]
X_train=X_dataset[800:]
X_val=X_dataset[400:800]
y_val=y_dataset[400:800]
y_test=y_dataset[:400]
y_train=y_dataset[800:] 
```

cluster 0 data를 이용한 linear regression

```python
lr = LinearRegression().fit(X_train, y_train)
y_train_hat = lr.predict(X_train)
print(y_train)
print(y_train_hat)
y_val_hat = lr.predict(X_val)
print(y_val)
print(y_val_hat)
```

```python
print('performance for TRAIN--------') 
print('train MAE : ', mean_absolute_error(y_train, y_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_hat))

print('performance for VAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_val_hat))
```

RMSE: 3683361255.831585, r_sqaure: 0.5861206380805186

P-value 값을 확인 하여 다중공산성의심되는 변수 찾아보기

```python
result = sm.OLS(y_train,X_train).fit()
print(result.summary())
```

다중공산성이 의심되는 p-value 값이 0.05보다 큰 7개의 컬럼('edu_b','traffic_b','t_household_b','work_20_30_40_50_num','work_10num_b ','apt_0.2b','apt_0.5b')에 대해 제거를 해보고 다시 regression 시도

```python
X_p_train=X_train.drop(columns=['edu_b','traffic_b','t_household_b','work_20_30_40_50_num','work_10num_b','apt_0.2b','apt_0.5b'])
X_p_val=X_val.drop(columns=['edu_b','traffic_b','t_household_b','work_20_30_40_50_num','work_10num_b','apt_0.2b','apt_0.5b'])
```

```python
lr = LinearRegression().fit(X_p_train, y_train)
y_p_train_hat = lr.predict(X_p_train)
print(y_train)
print(y_p_train_hat)
y_p_val_hat = lr.predict(X_p_val)
print(y_val)
print(y_p_val_hat)
print('performance for pTRAIN--------')
print('ptrain MAE : ', mean_absolute_error(y_train, y_p_train_hat))
print('ptrain RMSE : ', mean_squared_error(y_train, y_p_train_hat)**0.5)
print('ptrain R_square : ', r2_score(y_train, y_p_train_hat))

print('performance for pVAL--------')
print('p MAE : ', mean_absolute_error(y_val, y_p_val_hat))
print('p RMSE : ', mean_squared_error(y_val, y_p_val_hat)**0.5)
print('p R_square : ', r2_score(y_val, y_p_val_hat))
```

rmse:3693330981.3277903, r_square:0.5838771177723739

rmse는 증가했지만 r2의 값은 줄어 들었다 7개의 변수를 지운것이 효과가 없다.

vif값을 파악하여 다중공산성이 높은 변수를 제거해보기

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif["features"]= X_train.columns
```

```python
vif.round(1)
```

vif값이 10이상인 변수(총 20개)를 제거

```python
vifX = vif.iloc[vif['VIF Factor'].nlargest(20).index].features.values.tolist()
vifX
```

```python
X_vif_train=X_train.drop(columns=vifX)
X_vif_val=X_val.drop(columns=vifX)
```

```python
lr = LinearRegression().fit(X_vif_train, y_train)
y_vif_train_hat = lr.predict(X_vif_train)
print(y_train)
print(y_vif_train_hat)
y_vif_val_hat = lr.predict(X_vif_val)
print(y_val)
print(y_vif_val_hat)
print('performance for vifTRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_vif_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_vif_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_vif_train_hat))

print('performance for vifVAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_vif_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_vif_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_vif_val_hat))
```

rmse: 3986896358.539448, r_sqaure:0.5150968203670994

vif를 이용하여 변수 20개를  제거하였지만 rmse는 높아지고 r2은 낮아지므로 성능이 더 안좋아졌다.

pca(0.95정도)를 이용한 feature extraction

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
pca = PCA(n_components= 0.95) 
pca.fit(X_train_scaled)
X_train_pca = pca.transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
print("Original shape: {}".format(str(X_train_scaled.shape)))
print("Reduced shape: {}".format(str(X_train_pca.shape)))
```

38=>28 총 10개의 변수가 지워졌다.

```python
lr = LinearRegression().fit(X_train_pca, y_train)
y_train_pca_hat = lr.predict(X_train_pca)
print(y_train)
print(y_p_train_hat)
y_val_pca_hat = lr.predict(X_val_pca)
print(y_val)
print(y_val_pca_hat)
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_pca_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_pca_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_pca_hat))

print('performance for pcaVAL--------')
print('p MAE : ', mean_absolute_error(y_val, y_val_pca_hat))
print('p RMSE : ', mean_squared_error(y_val, y_val_pca_hat)**0.5)
print('p R_square : ', r2_score(y_val, y_val_pca_hat))
```

rmse:3724434604.6993384, r_square:0.5768387941966036

최종 모델 성능 확인하기 : 다중공산성을 줄이기 위해 vif 와 p-value를 이용하여 변수를 제거해보았지만 성능은 좋아지지 않았고, pca로 feature extraction을 시도해보았지만 이또한 성능은 좋아지지 않았다. 따라서 처음 데이터 그대로 linear regression을 적용한다.

```python
lr = LinearRegression().fit(X_train, y_train)
y_train_hat = lr.predict(X_train)
print(y_train)
print(y_train_hat)
y_test_hat = lr.predict(X_test)
print(y_test)
print(y_test_hat)
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_hat))

print('performance for test--------')
print('test MAE : ', mean_absolute_error(y_test, y_test_hat))
print('test RMSE : ', mean_squared_error(y_test, y_test_hat)**0.5)
print('test R_square : ', r2_score(y_test, y_test_hat))
```
