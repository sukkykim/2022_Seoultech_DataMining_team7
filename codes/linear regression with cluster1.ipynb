```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

cluster0 인 data를 test 와 val와  train으로 나누기
train : 17년도 1분기 ~ 21년도 2분기
val : 21년도 3분기
test : 21년도 4분기 

```python
dataset1=pd.read_csv('cluster_Label1.csv',index_col=0)
dataset1
```

```python
X_dataset=dataset1.drop(columns=['year','session','total_sale','s_zone','s_zone_code','s_zone_name'])
y_dataset=dataset1['total_sale']
X_test=X_dataset[:686]
X_train=X_dataset[1374:]
X_val=X_dataset[686:1374]
y_val=y_dataset[686:1374]
y_test=y_dataset[:686]
y_train=y_dataset[1374:]
```

cluster 0 data를 이용한 linear regression

```python
lr = LinearRegression().fit(X_train, y_train)
y_train_hat = lr.predict(X_train)
print(y_train)
print(y_train_hat)
y_val_hat = lr.predict(X_val)
print(y_val)
print(y_val_hat)
```

```python
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_hat))

print('performance for VAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_val_hat))
```

RMSE: 3893576259.683723, r_sqaure: 0.2829496626416472

P-value 값을 확인 하여 다중공산성의심되는 변수 찾아보기

```python
result = sm.OLS(y_train,X_train).fit()
print(result.summary())
```

다중공산성이 의심되는 p-value 값이 0.05보다 큰 9개의 컬럼('office','office_b','edu_b','traffic_b','t_household_b','apt_99','apt_0.2b','apt_0.6b','apt_avg_price')에 대해 제거를 해보고 다시 regression 시도

```python
X_p_train=X_train.drop(columns=['office','office_b','edu_b','traffic_b','t_household_b','apt_99','apt_0.2b','apt_0.6b','apt_avg_price'])
X_p_val=X_val.drop(columns=['office','office_b','edu_b','traffic_b','t_household_b','apt_99','apt_0.2b','apt_0.6b','apt_avg_price'])
```

```python
lr = LinearRegression().fit(X_p_train, y_train)
y_p_train_hat = lr.predict(X_p_train)
print(y_train)
print(y_p_train_hat)
y_p_val_hat = lr.predict(X_p_val)
print(y_val)
print(y_p_val_hat)
print('performance for pTRAIN--------')
print('ptrain MAE : ', mean_absolute_error(y_train, y_p_train_hat))
print('ptrain RMSE : ', mean_squared_error(y_train, y_p_train_hat)**0.5)
print('ptrain R_square : ', r2_score(y_train, y_p_train_hat))

print('performance for pVAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_p_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_p_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_p_val_hat))
```

rmse:  3885966206.0678306, r_square:0.2857498948488948

rmse는 감소했지만 r2의 값은 증가하였으므로 9개의 변수를 지운것이 효과가 있다.

p-value로 다중공산성이 의심되는 변수7개 가 제거된상태에서 vif값을 파악하여 다시한번 다중공산성이 높은 변수를 제거해보도록 하겠다.

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_p_train.values,i) for i in range(X_p_train.shape[1])]
vif["features"]= X_p_train.columns
```

```python
vif.round(1)
```

vif값이 10이상인 변수(총3개)를 제거

```python
vifX = vif.iloc[vif['VIF Factor'].nlargest(3).index].features.values.tolist()
vifX
```

```python
X_vif_train=X_p_train.drop(columns=vifX)
X_vif_val=X_p_val.drop(columns=vifX)
```

```python
lr = LinearRegression().fit(X_vif_train, y_train)
y_vif_train_hat = lr.predict(X_vif_train)
print(y_train)
print(y_vif_train_hat)
y_vif_val_hat = lr.predict(X_vif_val)
print(y_val)
print(y_vif_val_hat)
print('performance for vifTRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_vif_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_vif_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_vif_train_hat))

print('performance for vifVAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_vif_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_vif_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_vif_val_hat))
```

rmse:3895559410.170271, r_square:0.28221903312503416

vif를 이용하여 변수 3개를  추가 제거해보니 rmse는 커지고 r_sqaure은 작아지므로 성능은 좋아지지 않았다. 따라서 p-value로 9개의 변수를 제거한 데이터를 이용하겠습니다.

pca(0.95정도)를 이용한 feature extraction

```python
scaler = StandardScaler()
scaler.fit(X_p_train)
X_train_scaled = scaler.transform(X_p_train)
X_val_scaled = scaler.transform(X_p_val)
pca = PCA(n_components= 0.95) 
pca.fit(X_train_scaled)
X_train_pca = pca.transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
print("Original shape: {}".format(str(X_train_scaled.shape)))
print("Reduced shape: {}".format(str(X_train_pca.shape)))
```

29=>22 총 7개의 feature가 줄어들었다.

```python
lr = LinearRegression().fit(X_train_pca, y_train)
y_train_pca_hat = lr.predict(X_train_pca)
print(y_train)
print(y_p_train_hat)
y_val_pca_hat = lr.predict(X_val_pca)
print(y_val)
print(y_val_pca_hat)
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_pca_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_pca_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_pca_hat))

print('performance for pcaVAL--------')
print('p MAE : ', mean_absolute_error(y_val, y_val_pca_hat))
print('p RMSE : ', mean_squared_error(y_val, y_val_pca_hat)**0.5)
print('p R_square : ', r2_score(y_val, y_val_pca_hat))
```

rmse:3899153468.2389636, r_square:0.2808939671995526

최종 모델 성능 확인하기 : 다중공산성을 줄이기 위해 vif 와 p-value를 이용하여 변수 총 9개를 제거해보니 성능은 좋아졌으며, 그 후 pca로 feature extraction을 시도해보았지만 성능은 좋아지지 않았다. 따라서 p-value로  총 9개의 컬럼이 지워진 데이터로 test를 진행해보겠다.

```python
X_p_test=X_test.drop(columns=['office','office_b','edu_b','traffic_b','t_household_b','apt_99','apt_0.2b','apt_0.6b','apt_avg_price'])
lr = LinearRegression().fit(X_p_train, y_train)
y_train_hat = lr.predict(X_p_train)
print(y_train)
print(y_train_hat)
y_test_hat = lr.predict(X_p_test)
print(y_test)
print(y_test_hat)
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_hat))

print('performance for test--------')
print('test MAE : ', mean_absolute_error(y_test, y_test_hat))
print('test RMSE : ', mean_squared_error(y_test, y_test_hat)**0.5)
print('test R_square : ', r2_score(y_test, y_test_hat))
```
