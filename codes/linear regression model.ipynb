```python
import pandas as pd
```

dataset 불러오기

```python
dataset1=pd.read_excel('상관분석_매출통합 (1).xlsx',index_col=0)
dataset1
```

test data 와 train data 나누기
train data = 17년도 1분기 ~ 21년도 2분기
val data = 21년도 3분기
test data = 21년도 4분기

xdata= 전체 컬럼에서 session, s_zone, s_zone_code,s_zone_name, total_sale을 뺀 data,
ydata = totla_sale data  

```python
X_dataset=dataset1.drop(columns=['session','total_sale','s_zone','s_zone_code','s_zone_name'])
y_dataset=dataset1['total_sale']
```

```python
X_test=X_dataset[:1086]
X_train=X_dataset[2174:]
X_val=X_dataset[1086:2174]
y_val=y_dataset[1086:2174]
y_test=y_dataset[:1086]
y_train=y_dataset[2174:]
```

```python
print("X_train",X_train.shape)
print("y_train",y_train.shape)
print("X_val",X_val.shape)
print("y_val",y_val.shape)
print("X_test",X_test.shape)
print("y_test",y_test.shape)
```

linear regression model with raw data

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)
```

```python
y_train_hat = lr.predict(X_train)
print(y_train)
print(y_train_hat)
```

```python
y_val_hat = lr.predict(X_val)
print(y_val)
print(y_val_hat)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_hat))

print('performance for VAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_val_hat))
```

모델성능 파악 rmse : 3888471162.715774, val r_square:0.4730787755886081

다중공산성 판단을 위해 vif값 구하기

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif["features"]= X_train.columns
```

```python
vif.round(1)
```

vif 값이 10이상인 컬럼(t_household, income_range,medical_b, traffic_b,t_household_b,apt_complexes_num,apt_66_under, apt_66,99,132, apt_0.1b_under apt_0.1b,0.2b,0.3b,0.4b,0.5b,0.6b, apt_avg_width, total_living)을 제거하여 rmse, r_squared 변화 비교하기 

```python
vifX = vif.iloc[vif['VIF Factor'].nlargest(20).index].features.values.tolist()
vifX
```

```python
X_vif_train=X_train.drop(columns=vifX)
X_vif_val=X_val.drop(columns=vifX)
```

```python
lr = LinearRegression().fit(X_vif_train, y_train)
y_vif_train_hat = lr.predict(X_vif_train)
print(y_train)
print(y_vif_train_hat)
y_vif_val_hat = lr.predict(X_vif_val)
print(y_val)
print(y_vif_val_hat)
print('performance for vifTRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_vif_train_hat))
print('train RMSE : ', mean_squared_error(y_train, y_vif_train_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_vif_train_hat))

print('performance for vifVAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_vif_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_vif_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_vif_val_hat))
```

vif 10이상인 변수 제거후 모델 성능 파악 : rmse:4017940740.1925507, r_square:0.43740614847863246

vif가 10이상인 데이터 즉, 다중공선성이 높은 데이터를 제거하고 모델을 다시 만들어본결과 rmse 오차가 늘어나고 r_square값이 줄었으므로 변수를 제거하지 않고 그대로 가져가기로 한다.

p-value 값이 0.05를 넘어가면 다중공산성을 의심해볼만 함으로 p-value를 이용해 칼럼 제거 시도

```python
import statsmodels.api as sm
result = sm.OLS(y_train,X_train).fit()
print(result.summary())
```

확인해본결과 다중공산성 의심이 되는 p-value값이 0.05가 넘는 컬럼 총 9개('edu','edu_b','store_b','traffic_b','t_household_b','apt_66','apt_0.1b_under','apt_0.5b','apt_avg_price')를 지워서 validation으로 파악 해보기

```python
X_p_train=X_train.drop(columns=['edu','edu_b','store_b','traffic_b','t_household_b','apt_66','apt_0.1b_under','apt_0.5b','apt_avg_price'])
X_p_val=X_val.drop(columns=['edu','edu_b','store_b','traffic_b','t_household_b','apt_66','apt_0.1b_under','apt_0.5b','apt_avg_price'])
```

```python
lr = LinearRegression().fit(X_p_train, y_train)
y_p_train_hat = lr.predict(X_p_train)
print(y_train)
print(y_p_train_hat)
y_p_val_hat = lr.predict(X_p_val)
print(y_val)
print(y_p_val_hat)
print('performance for pTRAIN--------')
print('ptrain MAE : ', mean_absolute_error(y_train, y_p_train_hat))
print('ptrain RMSE : ', mean_squared_error(y_train, y_p_train_hat)**0.5)
print('ptrain R_square : ', r2_score(y_train, y_p_train_hat))

print('performance for pVAL--------')
print('val MAE : ', mean_absolute_error(y_val, y_p_val_hat))
print('val RMSE : ', mean_squared_error(y_val, y_p_val_hat)**0.5)
print('val R_square : ', r2_score(y_val, y_p_val_hat))
```

모델 성능 rmse:3897053072.9069104, r_sqaure:0.47075036399211334

RMSE와 r_square를 비교해본결과 전보다 오차가 커짐을 알 수 있었다. 다중공선성이 의심되는 변수들을 제거해봤지만 성능은 좋아지지 않았다.

PCA를 이용하여 feature extraction을 해보겠다.(0.95를 맞춰주는 칼럼뽑기)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
pca = PCA(n_components= 0.95) 
pca.fit(X_train_scaled)
X_train_pca = pca.transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
print("Original shape: {}".format(str(X_train_scaled.shape)))
print("Reduced shape: {}".format(str(X_train_pca.shape)))
```

38->27개로 총 11개의 feature가 줄어들었다.

```python
lr = LinearRegression().fit(X_train_pca, y_train)
y_train_pca_hat = lr.predict(X_train_pca)
print(y_train)
print(y_p_train_hat)
y_val_pca_hat = lr.predict(X_val_pca)
print(y_val)
print(y_val_pca_hat)
print('performance for TRAIN--------')
print('train MAE : ', mean_absolute_error(y_train, y_train_pca_hat))
print('train RMSE : ', mean_squared_error(y_train, y_train_pca_hat)**0.5)
print('train R_square : ', r2_score(y_train, y_train_pca_hat))

print('performance for pcaVAL--------')
print('p MAE : ', mean_absolute_error(y_val, y_val_pca_hat))
print('p RMSE : ', mean_squared_error(y_val, y_val_pca_hat)**0.5)
print('p R_square : ', r2_score(y_val, y_val_pca_hat))
```

모델성능 rmse:3918378208.4358964, r_square:0.46494228283712646

최종 모델 성능 확인하기 : 다중공산성을 줄이기 위해 vif 와 p-value를 이용하여 변수를 제거해보았지만 성능은 좋아지지 않았고, pca로 feature extraction을 시도해보았지만 이또한 성능은 좋아지지 않았다. 따라서 처음 데이터 그대로 linear regression을 적용한다.
